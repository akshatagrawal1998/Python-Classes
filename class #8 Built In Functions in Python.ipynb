```python
a = int(input("Enter a number to eb checked for palindrome\n"))
# reverse = original -> palindrome

# 354001
n = a
rev = 0
while(n!=0):
    r = n%10    # r=3  - r=5 , r=4
    #print(r)
    rev = rev*10+r  # rev = 3
    #print(rev)
    n = n//10   # n = 10045, n= 1004, n = 100
    #print(n)

if (a == rev):
    print(a, "is a palindrome number")    
else:
    print(a, "is not palindrome number")
```
